```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def hitung_psnr(gambar_asli, gambar_hasil):
    mse = np.mean((gambar_asli - gambar_hasil) ** 2)
    if mse == 0:
        return float('inf')  # PSNR tak terhingga jika citra sama persis
    maks_pixel = 255.0
    psnr = 20 * np.log10(maks_pixel / np.sqrt(mse))
    return psnr

# Ganti 'gambar_asli.jpg' dan 'gambar_hasil_stegano.png' dengan nama file gambar yang sesuai
nama_file_gambar_asli = '/content/as.jpg'
nama_file_gambar_hasil = '/content/as.jpg'

# Baca gambar asli dan hasil steganografi
gambar_asli = cv2.imread(nama_file_gambar_asli)
gambar_hasil = cv2.imread(nama_file_gambar_hasil)

# Periksa apakah gambar_hasil tidak kosong
if gambar_hasil is not None:
    # Tampilkan gambar_hasil
    plt.imshow(cv2.cvtColor(gambar_hasil, cv2.COLOR_BGR2RGB))
    plt.title('Gambar Hasil Steganografi')
    plt.show()

    # Konversi gambar ke ruang warna Grayscale jika perlu
    gambar_asli_gray = cv2.cvtColor(gambar_asli, cv2.COLOR_BGR2GRAY)
    gambar_hasil_gray = cv2.cvtColor(gambar_hasil, cv2.COLOR_BGR2GRAY)

    # Hitung PSNR
    psnr = hitung_psnr(gambar_asli_gray, gambar_hasil_gray)

    # Tampilkan hasil
    print(f"PSNR antara gambar asli dan hasil steganografi: {psnr} dB")
else:
    print(f"Error: Gagal membaca gambar hasil steganografi.")
```
